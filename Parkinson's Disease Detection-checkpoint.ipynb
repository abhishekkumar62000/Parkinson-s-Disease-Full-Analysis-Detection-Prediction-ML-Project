### Parkinson's Disease Symptons Overview:----
Parkinson's disease is a brain disorder that causes movement problems, mental health issues, and other health problems: 
Movement issues: Tremors, stiffness, slowness of movement, poor balance and coordination, difficulty walking and talking 
Mental health issues: Depression, anxiety, apathy 
Other health issues: Sleep problems, memory difficulties, fatigue, bladder problems, constipation, blood pressure changes, loss of smell, pain, sexual symptoms 
Parkinson's disease is caused by nerve cells in the brain not producing enough dopamine. Symptoms usually begin gradually, often on one side of the body, and worsen over time. There is no cure, but treatments can help relieve symptoms. 
Some risk factors for Parkinson's disease include:
Age: Most people with Parkinson's first develop the disease after age 60, but about 5% to 10% experience onset before the age of 50 
Family history: People with a family history of the disease have a higher risk 
Exposure to chemicals: Exposure to air pollution, pesticides, and solvents may increase risk 
Some lifestyle changes may help reduce the risk of Parkinson's disease, including: Exercising regularly and Eating foods that contain turmeric and flavonoids.

```python
X=data.drop(columns = ['name', 'status'], axis=1)
y=data['status'] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
y_train
```

```python
# Data Standardization
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
train_acc = accuracy_score(y_train, pred_train)
```

```python
train_acc
```

```python
pred_test = model.predict(X_test)
```

```python
test_acc = accuracy_score(y_test, pred_test)
```
